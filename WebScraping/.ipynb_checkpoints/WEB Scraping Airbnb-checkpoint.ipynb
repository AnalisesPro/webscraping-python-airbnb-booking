# Imports

```python
%%time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from selenium.webdriver import FirefoxProfile
import pandas as pd
import xlrd
import os

import warnings
warnings.filterwarnings('ignore')

# Rodar em segundo plano
os.environ['MOZ_HEADLESS'] = '1'
```

# Configurações

```python
%%time
firefox = r'DriversSelenium\geckodriver.exe'
options = FirefoxOptions()
options.add_argument('--incognito')

planilha = r'CrawlerFresteCB_Links.xlsx'
```

# Carrega configurações da pesquisa

```python
links = pd.read_excel(
        open(planilha, 'rb'),
        sheet_name='Links', engine='openpyxl')

ceps = pd.read_excel(
        open(planilha, 'rb'),
        sheet_name='Ceps', engine='openpyxl')


ceps['cep'] = ceps['cep'].astype('str')
ceps['cep'] = ceps['cep'].str.replace('-', '')

qtdSkus = len(links)
qtdCeps = len(ceps)
combinacoes = qtdSkus*qtdCeps

#print(links, '\n\n', ceps, '\n')
print('Qtd skus:', qtdSkus)
print('Qtd ceps:', qtdCeps)
print('Combinações:', combinacoes)
```

# Recupera URLs dos SKUs

```python
%%time

#links = ['https://www.casasbahia.com.br/1001298751/b', \
#          'https://www.casasbahia.com.br/1001290216/b', \
#          'https://www.casasbahia.com.br/55008110/b']

linksSkus = []
count = 0

for link in links.values:
    #print(link[1])
    try:
        # Abre Browser
        driver = webdriver.Firefox(options=options, executable_path=firefox)
        driver.maximize_window()
        wait = WebDriverWait(driver, 5)

        # Acessa Site Casas Bahia
        driver.get(link[1])

        # Recupera elemento com o link do sku
        linkSite = wait.until(EC.element_to_be_clickable((By.XPATH,\
                        '//*[@id="__next"]/div[2]/div/div/div[3]/div[2]/div/section/ul/li/div/a'))).get_attribute('href')
        
        # Salva numa lista
        linksSkus.append([linkSite, link[0]])
    except:  
        linksSkus.append('erro', link[0])
        driver.quit()
        count = count + 1
        continue
        
        
    driver.quit()
    count = count + 1
    
    #print(count, 'de', qtdSkus, '-----', round((count/qtdSkus)*100, 2), '%')
```

```python
linksSkus = pd.DataFrame(linksSkus, columns=['link', 'sku'])
linksSkus.link[4]
```

# Recupera valores de Frete

```python
%%time

#ceps = ['07718035', '66917140']

result_raspagem = []
count = 0

for link in linksSkus.values:
    url_contexto = link[0]
    count = count + len(ceps) - 1
    for cep in ceps.values:
        try:
            # Abre Browser
            driver = webdriver.Firefox(options=options, executable_path=firefox)
            driver.maximize_window()
            wait = WebDriverWait(driver, 5)

            # Acessa Site Casas Bahia
            driver.get(url_contexto)

            # Digita Frete
            wait.until(EC.element_to_be_clickable((By.ID, 'frete'))).send_keys(cep[0])

            # Captura valor Frete
            valor_frete = wait.until(EC.element_to_be_clickable((By.XPATH,\
                            '/html/body/div[1]/div/div[2]/div[2]/div[2]/div[7]/div/div/div/div[2]/div'))).text

            # Salva numa lista
            result_raspagem.append([link[1], cep[0], valor_frete])
            count = count - 1
        except:  
            result_raspagem.append([link[1], cep[0], 'erro'])
            driver.quit()
            #print(url_contexto, '\n')
            break
        driver.quit()
        
        count = count + 1
        
    print(count, 'de', combinacoes, '-----', round((count/combinacoes)*100, 2), '%')
```

# Trata os dados recuperados

```python
df = pd.DataFrame(result_raspagem, columns=['idsku', 'cep', 'raspagem'])
df.to_csv(r'raspagem_bruta_2.csv', sep=';', encoding='utf-8-sig', decimal=',', index=False)
df_copia = df.copy()
df
```

```python
dados_brutos = list(df_copia.raspagem.str.split('Normal\n').str[1].str.split('\n').str[1])
df2 = pd.DataFrame(dados_brutos, columns=['Normal_valor'])

dados_brutos = list(df_copia.raspagem.str.split('Retira Rápido\n').str[1].str.split('\n').str[1])
df2['RetiraRapido_valor'] = dados_brutos

dados_brutos = list(df_copia.raspagem.str.split('Expressa\n').str[1].str.split('\n').str[1])
df2['Expressa_valor'] = dados_brutos

dados_brutos = list(df_copia.raspagem.str.split('Agendada\n').str[1].str.split('\n').str[1])
df2['Agendada_valor'] = dados_brutos
```

```python
# Remove espacos em branco
df2.Normal_valor = df2.Normal_valor.str.replace(' ', '')
df2.RetiraRapido_valor = df2.RetiraRapido_valor.str.replace(' ', '')
df2.Expressa_valor = df2.Expressa_valor.str.replace(' ', '')
df2.Agendada_valor = df2.Agendada_valor.str.replace(' ', '')
```

```python
# Atribui zero ao frete gratis
df2.Normal_valor = df2.Normal_valor.str.replace('Grátis', '000')
df2.RetiraRapido_valor = df2.RetiraRapido_valor.str.replace('Grátis', '000')
df2.Expressa_valor = df2.Expressa_valor.str.replace('Grátis', '000')
df2.Agendada_valor = df2.Agendada_valor.str.replace('Grátis', '000')
```

```python
# Remove o R$
df2.Normal_valor = df2.Normal_valor.str[2:]
df2.RetiraRapido_valor = df2.RetiraRapido_valor.str[2:]
df2.Expressa_valor = df2.Expressa_valor.str[2:]
df2.Agendada_valor = df2.Agendada_valor.str[2:]
```

```python
# Remove espacos em branco e substitui virgula por ponto
df2.Normal_valor = df2.Normal_valor.str.replace(',','.')
df2.RetiraRapido_valor = df2.RetiraRapido_valor.str.replace(',','.')
df2.Expressa_valor = df2.Expressa_valor.str.replace(',','.')
df2.Agendada_valor = df2.Agendada_valor.str.replace(',','.')
```

```python
# Adiciona colunas ao df
df['Normal_valor'] = df2.Normal_valor.values
df['RetiraRapido_valor'] = df2.RetiraRapido_valor.values
df['Expressa_valor'] = df2.Expressa_valor.values
df['Agendada_valor'] = df2.Agendada_valor.values

# Transforma tipagem dos dados
df.idsku = df.idsku.astype('int64')
df.cep = df.cep.astype('int64')
df.Normal_valor = df.Normal_valor.astype(float)
df.RetiraRapido_valor = df.RetiraRapido_valor.astype(float)
df.Expressa_valor = df.Expressa_valor.astype(float)
df.Agendada_valor = df.Agendada_valor.astype(float)

# Remove coluna raspagem
df.drop(columns=['raspagem'], inplace=True)
```

# Base final com Frete Normal Casas Bahia

```python
df.to_csv(r'raspagem_20213012_final_2.csv', sep=';', encoding='utf-8-sig', decimal=',', index=False)
df
```
